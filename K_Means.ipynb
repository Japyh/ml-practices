```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X, _ = make_blobs(n_samples=1000, centers=4, random_state=42, cluster_std=1.5)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c='red', marker='o', edgecolor='black')
plt.title('Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, marker='o', edgecolor='black')
plt.title('Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
centers = kmeans.cluster_centers_
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', edgecolor='black')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', edgecolor='black', s=200)
plt.title('Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
