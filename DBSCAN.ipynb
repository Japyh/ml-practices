```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
X, _ = make_circles(n_samples=1500, factor=0.5, noise=0.08, random_state=42)
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c='red', marker='o', edgecolor='black')
plt.title('Circles')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.11, min_samples=5)
clusters = dbscan.fit_predict(X)
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=clusters, marker='o', edgecolor='black', cmap='viridis')
plt.title('Circles')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
