```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
X, _ = make_blobs(n_samples=1500, centers=4, cluster_std=1.5, random_state=42)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c='red', marker='o', edgecolor='black')
plt.title('Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
linkage_methods = ['ward', 'complete', 'average', 'single']

plt.figure(figsize=(20, 10))
for i, method in enumerate(linkage_methods, 1):
    
    model = AgglomerativeClustering(n_clusters=4, linkage=method)
    clusters_labels = model.fit_predict(X)

    plt.subplot(2, 4, i)
    plt.title(f'{method.capitalize()} Linkage Dendrogram')
    dendrogram(linkage(X, method=method), no_labels=True)
    plt.xlabel('Sample Index')
    plt.ylabel('Distance')
    
    plt.subplot(2, 4, i + 4)
    plt.scatter(X[:, 0], X[:, 1], c=clusters_labels, cmap='viridis')
    plt.title(f'{method} Linkage Clusters')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.tight_layout()
```
